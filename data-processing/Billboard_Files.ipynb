### Billboard Files
#### Used to merge the 2 billboard files into a single source
- This is just for preprocessing the files and should not have to be rerun
- The csv output will be used to fetch more lyrics

```python
import pandas as pd
import numpy as np
```

```python
#these are the orginal files. note: they do not contain any of the scarped lyrics
lyrics_file = 'data/originals/billboard_lyrics_1964-2015.csv'
songs_file = 'data/originals/billboard_hot_100_1960_to_2018.csv'
```

```python
lyrics_data = pd.read_csv(lyrics_file)
songs_data = pd.read_csv(songs_file)
```

```python
songs_data.head()
```

```python
lyrics_data.head()
```

```python
#merge the file
song_merged = pd.merge(songs_data, lyrics_data,  how='left', left_on = ['Rank','Year'], right_on = ['Rank','Year']).set_index("RankId")
```

```python
song_merged.head()
```

```python
#export to csv to use for processing lyrics
song_merged.to_csv('data/billboard_songs_merged.csv')
```
